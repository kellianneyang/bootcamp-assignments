## Task

- Recreate this histogram using seaborn.
- Annotate the minimum recommend BMI with a red line at 18.5.
- Annotate the maximum recommended BMI with a red line at 24.9.

Download your notebook and submit the file here.



```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# recreate the histogram with seaborn

# sns.set(style = 'darkgrid')
ax = sns.histplot(data = df, x = 'bmi')

# set title: 'bmi'
ax.set_title('bmi')

# add red line: 'Min Recommended BMI: 18.5'
ax.axvline(18.5, 
           color = 'red', 
           label = 'Min Recommended BMI: 18.5')

# add red line: 'Max Recommended BMI: 24.9'
ax.axvline(24.9, 
           color = 'red', 
           label = 'Max Recommended BMI: 24.9')

# add legend
ax.legend(loc = 'upper right',
          fontsize = 7)

plt.show();
```
