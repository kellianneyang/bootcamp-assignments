3. Select just data where the 'Change 2019 -2020' column is greater than 50% (greater than .5).


```python
change_filter = bike_df['Change 2019-2020'] >= .5
print(type(change_filter))
display(change_filter)
```

4. Put these together! When filtering you can use "&" to combine two filters. Determine how many weeks in the second half of the year (after July 31, 2020) also had a change greater than 50%? Hint: use len().


```python
# apply df_filtered3 (weeks after week 32) AND change_filter (2019-2020 change greater than .5)
two_filters = bike_df.loc[df_filtered3 & change_filter, :]
display(two_filters)
print(f"{len(two_filters)} weeks in the second half of the year also had a change greater than 50%.")
```

5. Next, make a new combined filter where you only include changes greater than 100% that also occurred after July 31, 2020. Display the filtered DataFrame.
